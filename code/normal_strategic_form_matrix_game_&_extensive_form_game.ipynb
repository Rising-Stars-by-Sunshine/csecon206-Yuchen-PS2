the Chicken game is a model of conflict between risk-taking and risk-avoidance of two players. One of the classical game scenarios is when two drivers drive toward each other, each can choose to swerve or straight. If one of them chooses to swerve and the other chooses to straight, the former one will be seen as chickening out and will receive a lower payoff (2), while the latter one will receive a higher payoff (7) since he’s braver. If both choose to swerve, it will be a tie, and both receive a moderate payoff (6). While, if both players choose to straight, they will crash into each other, and receives no payoff (0), since they fail to avoid the risk.

```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[0, 7],
        [2, 6]]) # A is the row player

B = np.array([[0, 2],
        [7, 6]]) # B is the column player
     
```

```python
# Form the game
game2 = nash.Game(A,B)
game2
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game2.support_enumeration()
for eq in equilibria:
    print(eq)
```

In this game, there are three Nash equilibrium: 

1.Player A swerves, and Player B goes straight. The reason for this choice is that player A wants to avoid the danger of a crash, while Player B wants to maximize revenue. In this case, payoff is 2 and 7.

2.Player A goes straight and player B swerves. In this case, payoff is 7 or 2.

In both equilibriums, where one player chooses not to risk and the second player chooses the high payoff, both players get a satisfactory outcome, so they don't change strategies, so the outcome is stable. Therefore, they are pure strtagies.

3.This is a mixed strategies. The calculation process is: Assume that player A swerves with probability p and goes straight with probability 1-p, B chooses the options with probability p and 1-p due to the symmetry of the game. If B chooses to swerve, then the expected payoff for A is 6(1-p)+2p. If B chooses to go straight, then the expected payoff for A is 7(1-p)+0. Setting the expected payoffs for player A's strategies equal to each other, then we have: 6(1-p)+2p=7(1-p)+0. By calculation, p=1/3 and 1-p=2/3.

**Extensive form game: Ultimatum game**

The rule of the ultimatum game is that one player, the proposer has some endowment. The proposer then needs to split the money with another player, the responder, who knows the total amount of the money. Then the responder can accept the proposal or reject it. If the responder accepts, the payoff is the same as the proposal. However, if the responder rejects, both receive nothing. And both players know the rule of the game in advance.
